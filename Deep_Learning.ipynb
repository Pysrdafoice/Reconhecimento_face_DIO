```python
from numpy import loadtxt
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = loadtxt('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv', delimiter= ',')
```

```python
x =dataset[:,0:8]
y = dataset[:,8]
```

```python
#definiro modelo keras
model = Sequential()

model.add(Dense(12, input_dim = 8, activation= 'relu')) # Changed 'input_din' to 'input_dim'
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x, y, epochs=150, batch_size=10)
```

```python
# Substitua o obsoleto 'predict_classes' por 'predict' e argmax para previsão de classe
_, accuracy = model.evaluate(x, y)
print('Acuracia: %2.f' % (accuracy*100))

predictions = model.predict(x)
predicted_classes = np.argmax(predictions, axis=1)  # Obtenha a classe com maior probabilidade
for i in range(5):
  print("%s => %d (esperado %d)" % (x[i] , predicted_classes[i], y[i]))
  print(predictions[i])
  print()
```
